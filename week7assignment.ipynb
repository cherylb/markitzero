```python
# -*- coding: utf-8 -*-
"""
Created on Sun Jul 20 15:53:02 2014

@author: Cheryl
"""

from bs4 import BeautifulSoup
import urllib

while True:
    myurl = raw_input('Enter url>>')
    # myurl = "http://www.revolutionanalytics.com/r-language-resources"
    print(myurl)

    #checking url
    if len(myurl)<1:
        break

    try:
        html = urllib.urlopen(myurl).read()
    except:
            print("could not read url")
            break

    #parse url, create list of tags
    soup = BeautifulSoup(html)
    tags = soup('a')

    print('Number of links:', len(tags))
    for tag in tags:
        print(tag.get('href', None))
    break
```
